# Minimal example of drawing ROC curve with the turtle

## Training a random classifier on some dummy data

```python
from lib.drawing import draw_roc_curve_with_a_turtle

import numpy as np
from sklearn.metrics import RocCurveDisplay
```

```python
y_true = [0] * 100 + [1] * 100
y_pred = np.random.random(len(y_true))
```

## Draw with sklearn

```python
roc_display = RocCurveDisplay.from_predictions(
    y_true,
    y_pred,
    name=f"Random scores",
    color="blue",
    plot_chance_level=True,
)
_ = roc_display.ax_.set(
    xlabel="False Positive Rate",
    ylabel="True Positive Rate",
    title="ROC curve: Random classifier",
)
```

## Draw with a turtle

```python
canvas = draw_roc_curve_with_a_turtle(y_true, y_pred, "blue", stop_at=1)
```
