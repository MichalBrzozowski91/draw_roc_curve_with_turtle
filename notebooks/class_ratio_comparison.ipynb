# Class Ratio Comparison

This notebook demonstrates how different class ratios affect the visual representation of ROC curves drawn with turtle graphics.

The class ratio affects the "grid" because step sizes are calculated as `canvas_size / class_count`:
- More positives = smaller vertical steps
- More negatives = smaller horizontal steps

## Balanced (50:50) - 100 positives, 100 negatives

With equal class counts, the grid is square.

```python
y_true_balanced = [0] * 100 + [1] * 100
y_pred_balanced = np.random.random(len(y_true_balanced))

draw_roc_curve_with_a_turtle(y_true_balanced, y_pred_balanced)
```

## Positive-heavy (90:10) - 180 positives, 20 negatives

With many more positives, vertical steps are small and horizontal steps are large, creating a tall narrow grid.

```python
y_true_pos_heavy = [0] * 20 + [1] * 180
y_pred_pos_heavy = np.random.random(len(y_true_pos_heavy))

draw_roc_curve_with_a_turtle(y_true_pos_heavy, y_pred_pos_heavy)
```

## Negative-heavy (10:90) - 20 positives, 180 negatives

With many more negatives, horizontal steps are small and vertical steps are large, creating a wide short grid.

```python
y_true_neg_heavy = [0] * 180 + [1] * 20
y_pred_neg_heavy = np.random.random(len(y_true_neg_heavy))

draw_roc_curve_with_a_turtle(y_true_neg_heavy, y_pred_neg_heavy)
```

## Conclusion

Despite the visual differences in grid shape, all three examples show approximately 0.5 AUC since they use a random classifier. The diagonal reference line always represents random performance regardless of class ratio.
